```python
from pymongo import MongoClient
import pprint
connectionstring="mongodb+srv://Clover909:Clover909@cluster1.xvd5b.mongodb.net/?retryWrites=true&w=majority&appName=Cluster1"
client=MongoClient(connectionstring)
db=client["sample_mflix"]
collection=db["movies"]
```

```python
collection.find_one({"genres":"Action"})
for x in collection.find({"year":{"$gt":2000}}).limit(5):
    pprint.pprint(x)
for x in collection.find({"imdb.rating":{"$gt":8.5}}).limit(5):
    pprint.pprint(x)
collection.find({"genres":{"$all":["Adventure","Action"]}}).limit(5)
```

```python
for x in collection.find({"genres":"Comedy"}).sort("imdb.rating",-1).limit(5):
    pprint.pprint(x)
for x in collection.find({"genres":"Drama"}).sort("year",1).limit(5):
    pprint.pprint(x)
```

```python
aggregate1=[{"$unwind":"$genres"},{"$group":{"_id":"$genres","avg_rating":{"$avg":"$imdb.rating"}}},{"$sort":{"avg_rating":-1}},{"$limit":5}]
for x in collection.aggregate(aggregate1):
    pprint.pprint(x["_id"])
aggregate2=[{"$unwind":"$directors"},{"$group":{"_id":"$directors","avg_rating":{"$avg":"$imdb.rating"}}},{"$sort":{"avg_rating":-1}},{"$limit":5}]
collection.aggregate(aggregate2)
aggregate3=[{"$group":{"_id":"$year","total":{"$sum":1}}},{"$sort":{"_id":1}}]
collection.aggregate(aggregate3)
```

```python
collection.update_one({"title":"The Godfather"},{"$set":{"imdb.rating":9.5}})
collection.update_many({"genres":"Horror","imdb.rating":""},{"$set":{"imdb.rating":6.0}})
collection.delete_many({"year":{"$lt":1950}})
```

```python
collection.drop_indexes()
collection.create_index([("title","text")])
collection.find({"$text":{"$search":"love"}})
```

```python
collection.drop_indexes()
collection.create_index([("title","text"),("plot","text")])
for x in collection.find({"$text":{"$search":"war"}}).sort("imdb.rating",-1).limit(5):
    pprint.pprint(x)
```
